```python
#1.质因数分解函数
```

```python
def prime_factor(n):
    factors=[]
    i=2
    while i*i<=n:
        if n%i:
            i+=1
        else:
            n//=i
            factors.append(i)
    if n>i:
        factors.append(n)
    return factors

print("90的质因数分解结果:",prime_factor(90))
print("100的质因数分解结果：",prime_factor(100))
```

```python
# 3.矩阵和线性方程组
```

```python
A=np.array([[1,1,0],[-4,3,0],[1,0,2]])
b=np.array([0,1,0])

eigenvalues,eigenvectors=np.linalg.eig(A)
print("特征值：",eigenvalues)
print("特征向量：",eigenvectors)

x=np.linalg.solve(A,b)
print("线性方程组的解：",x)
```

```python
# 4.绘制正弦曲线
```

```python
x=np.linspace(0,2*np.pi,1000)
y=np.sin(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('siin(x)')
plt.title('正弦函数曲线')
plt.show()
```

```python
# 5.定义数位求和函数
```

```python
def THU(n):
    while n>=10:
        n=sum(int(digit)for digit in str(n))
    return(n)

print("输入29的结果：",THU(29))
```
